# Step 1

## Todo

Write a function `compute_ratio`:

1. take as argument a list of $N$ numbers. These numbers are non negative.
2. return another list of length $N-1$
3. the $k^{th}$ element of the returned list is the ratio of the $k^{th}$ and $(k+1)^{th}$ number of the argument list

## Test

```
print(compute_ratio([]))
print(compute_ratio([8]))
print(compute_ratio([3j-9, 3j+9, 3]))
print(compute_ratio([5, 5, 50, 2.5]))
```

should return:

```
[]
[]
[(-0.8+0.6j), (3+1j)]
[1.0, 0.1, 20.0]
```
