Plot the following data: 

- line: dashed, blue
- markers: squares, red

Save the figure.

```python
x = [1, 2, 3]
y = [5, 7, 8]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(x, y, '-.b')
ax.plot(x, y, 'sr')

fig.savefig('new_figure.png', dpi=200, bbox_inches='tight')
```
